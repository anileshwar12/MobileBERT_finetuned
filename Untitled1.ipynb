```python
import torch
from transformers import MobileBertForQuestionAnswering, AutoTokenizer, Trainer, TrainingArguments
from datasets import load_dataset
import wandb
```

```python
dataset = load_dataset("squad")
```

```python
model_checkpoint = "google/mobilebert-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
model = MobileBertForQuestionAnswering.from_pretrained(model_checkpoint)
```

```python
def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]

    inputs = tokenizer(
        questions,
        examples["context"],
        truncation=True,
        padding="max_length",
        max_length=384,
        return_tensors="pt"
    )

    start_positions = []
    end_positions = []

    for i in range(len(examples["answers"])):
        # Check if the example has an answer
        if len(examples["answers"][i]["text"]) == 0:
            start_positions.append(0)  # Default value
            end_positions.append(0)
            continue

        answer_start = examples["answers"][i]["answer_start"][0]
        answer_text = examples["answers"][i]["text"][0]
        answer_end = answer_start + len(answer_text)

        offsets = inputs.encodings[i].offsets

        # Find token indices
        token_start_index = next((idx for idx, (s, e) in enumerate(offsets) if s >= answer_start), 0)
        token_end_index = next((idx for idx, (s, e) in enumerate(offsets) if e >= answer_end), 0)

        start_positions.append(token_start_index)
        end_positions.append(token_end_index)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions

    return inputs
```

```python
tokenized_datasets = dataset.map(preprocess_function, batched=True)
```

```python
training_args = TrainingArguments(
    output_dir="./mobilebert-finetuned",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=50,
    push_to_hub=False
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["validation"],
)
```

```python
trainer.train()
```

```python
model.save_pretrained("mobilebert-finetuned")
tokenizer.save_pretrained("mobilebert-finetuned")
```
